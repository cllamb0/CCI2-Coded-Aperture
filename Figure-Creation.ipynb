```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
results = pd.read_csv('Optimized-Seeds/seed-metadata-with-results.csv')
```

```python
results
```

## Sensitivity Metric 1

```python
ffs = [0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7]
sw = 1

plt.figure(dpi=300)
for i, ff in enumerate(ffs):
    temp_result = results[(results['Fill Fraction'] == ff) & (results['Sensitivity Weight'] == sw)]
    if i == 0:
        plt.scatter(ff, temp_result['final_metricF'].values.mean(), color='red', label='Magnification 4')
        plt.scatter(ff, temp_result['final_metric3'].values.mean(), color='green', label='Magnification 3')
        plt.scatter(ff, temp_result['final_metric2'].values.mean(), color='blue', label='Magnification 2')
    else:
        plt.scatter(ff, temp_result['final_metricF'].values.mean(), color='red')
        plt.scatter(ff, temp_result['final_metric3'].values.mean(), color='green')
        plt.scatter(ff, temp_result['final_metric2'].values.mean(), color='blue')
plt.xlabel('Fill Fraction')
plt.ylabel('Combined Metric')
plt.legend()
plt.show()
```
